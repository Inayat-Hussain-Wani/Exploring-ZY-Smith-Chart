```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cmath
import ipywidgets as widgets
from IPython.display import display
```

```python
# normalized impedance mapping on gamma plane
arr1 = np.round(np.linspace(0, 1, 10), 1)
arr2 = np.round(np.linspace(1, 2, 5), 1)
arr3 = np.array([3, 4, 5, 10, 20, 50])
zreal_values = np.concatenate((arr1, arr2, arr3))
zimag_values = np.concatenate((-zreal_values[::-1], zreal_values))
points = []  # List to store points
# Function to plot points
def point(zreal=0, zimag=0):
    z = zreal + zimag * 1j
    gamma = (z - 1) / (z + 1)
    x = gamma.real
    y = gamma.imag
    points.append((x, y))
    
    plt.figure()
    for pt in points:
        plt.scatter(pt[0], pt[1], linewidth=2, color='b')
    
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.grid(True)
    plt.xlabel("Real Part")
    plt.ylabel("Imaginary Part")
    plt.title("Gamma Plane")
    plt.show()

# Function to clear the plot
def clear_plot(button):
    global points
    points = []  # Clear the points
    plt.clf()  # Clear the current figure
    

# Clear button
clear_button = widgets.Button(description="Clear Plot")
clear_button.on_click(clear_plot)

# Sliders
zreal_slider = widgets.SelectionSlider(options=zreal_values, description="zreal")
zimag_slider = widgets.SelectionSlider(options=zimag_values, description="zimag")

interactive_plot = widgets.interactive_output(point, {"zreal": zreal_slider, "zimag": zimag_slider})
controls = widgets.HBox([widgets.VBox([zreal_slider, zimag_slider]), clear_button])
widgets.VBox([interactive_plot, controls])
```

```python
# Define data for constant conductance and susceptance curves
arr1 = np.linspace(0, 1, 10)
arr2 = np.linspace(1, 2, 5)
arr3 = np.array([3, 4, 5, 10, 20, 50])
yreal_temp = np.concatenate((arr1, arr2, arr3))
yimag = np.linspace(-50, 50, 2000) * 1j

# Plot constant-conductance circles
for yreal in yreal_temp:
    y = yreal + yimag
    gamma = -(y - 1) / (y + 1)
    plt.plot(gamma.real, gamma.imag, linewidth=1, color='black')

# Data for constant susceptance lines
yreal = np.linspace(0, 50, 2000)
yimag_inductor = -np.concatenate((arr1, arr2, arr3)) * 1j
yimag_capacitor = np.concatenate((arr1, arr2, arr3)) * 1j
yimag_temp = np.concatenate((yimag_inductor, yimag_capacitor))

# Plot constant-susceptance circles
for yimag in yimag_temp:
    y = yreal + yimag
    gamma = -(y - 1) / (y + 1)
    plt.plot(gamma.real, gamma.imag, linewidth=1, color='red')

# Configure plot
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.title("Gamma Plane - Y-Smith Chart")
plt.show()
```

```python
# Both side-by-side
# Set figure size and create subplots
fig_size = (10, 5)
fig, axs = plt.subplots(1, 2, figsize=fig_size)

# Generate common data
arr1 = np.linspace(0, 1, 10)
arr2 = np.linspace(1, 2, 5)
arr3 = np.array([3, 4, 5, 10, 20, 50])
combined_real = np.concatenate((arr1, arr2, arr3))
line_imaginary = np.linspace(-50, 50, 2000) * 1j

# Z-Smith chart
zreal = combined_real
zimag = line_imaginary

for zreal_val in zreal:
    z = zreal_val + zimag
    gamma = (z - 1) / (z + 1)
    axs[0].plot(gamma.real, gamma.imag, linewidth=1, color='black')

zimag_inductor = combined_real * 1j
zimag_capacitor = -combined_real * 1j
zimag_combined = np.concatenate((zimag_inductor, zimag_capacitor))

for zimag_val in zimag_combined:
    z = np.linspace(0, 50, 2000) + zimag_val
    gamma = (z - 1) / (z + 1)
    axs[0].plot(gamma.real, gamma.imag, linewidth=1, color='red')

axs[0].set_xlim(-1, 1)
axs[0].set_ylim(-1, 1)
axs[0].grid(True)
axs[0].set_xlabel("Real Part")
axs[0].set_ylabel("Imaginary Part")
axs[0].set_title("Gamma Plane - Z-Smith Chart")

# Y-Smith chart
yreal = combined_real
yimag = line_imaginary

for yreal_val in yreal:
    y = yreal_val + yimag
    gamma = -(y - 1) / (y + 1)
    axs[1].plot(gamma.real, gamma.imag, linewidth=1, color='black')

yimag_inductor = -combined_real * 1j
yimag_capacitor = combined_real * 1j
yimag_combined = np.concatenate((yimag_inductor, yimag_capacitor))

for yimag_val in yimag_combined:
    y = np.linspace(0, 50, 2000) + yimag_val
    gamma = -(y - 1) / (y + 1)
    axs[1].plot(gamma.real, gamma.imag, linewidth=1, color='red')

axs[1].set_xlim(-1, 1)
axs[1].set_ylim(-1, 1)
axs[1].grid(True)
axs[1].set_xlabel("Real Part")
axs[1].set_ylabel("Imaginary Part")
axs[1].set_title("Gamma Plane - Y-Smith Chart")

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

```python
#Complete ZY-Smith chart
# Generate real and imaginary components
arr1 = np.linspace(0, 1, 4)
arr2 = np.array([2, 5, 10, 20, 50])

# Z-real and Z-imaginary components
zreal = np.concatenate((arr1, arr2))
zimag = np.linspace(-50, 50, 2000) * 1j

# Plot constant-resistance circles
for zr in zreal:
    z = zr + zimag
    gamma = (z - 1) / (z + 1)
    plt.plot(gamma.real, gamma.imag, linewidth=1, color='red', label='Resistance Circle' if zr == zreal[0] else "")

# Z-imaginary components for inductors and capacitors
zinductor = np.concatenate((arr1, arr2)) * 1j
zcapacitor = -np.concatenate((arr1, arr2)) * 1j
zreal_line = np.linspace(0, 50, 2000)

# Plot constant-reactance circles
for zi in np.concatenate((zinductor, zcapacitor)):
    z = zreal_line + zi
    gamma = (z - 1) / (z + 1)
    plt.plot(gamma.real, gamma.imag, linewidth=1, color='blue', label='Reactance Circle' if zi == zinductor[0] else "")

# Y-real and Y-imaginary components
yreal = np.concatenate((arr1, arr2))
yimag = np.linspace(-50, 50, 2000) * 1j

# Plot constant-conductance circles
for yr in yreal:
    y = yr + yimag
    gamma = -(y - 1) / (y + 1)
    plt.plot(gamma.real, gamma.imag, linewidth=1, color='black', label='Conductance Circle' if yr == yreal[0] else "")

# Y-imaginary components for inductors and capacitors
yinductor = -np.concatenate((arr1, arr2)) * 1j
ycapacitor = np.concatenate((arr1, arr2)) * 1j

# Plot constant-susceptance circles
for yi in np.concatenate((yinductor, ycapacitor)):
    y = zreal_line + yi
    gamma = -(y - 1) / (y + 1)
    plt.plot(gamma.real, gamma.imag, linewidth=1, color='green', label='Susceptance Circle' if yi == yinductor[0] else "")

# Chart setup
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.title("Complete ZY-Smith Chart")
plt.show()


```
